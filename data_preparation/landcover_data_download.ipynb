# Downloader from bounding boxes

```python
import ee #pip install earthengine-api
import geemap.core as geemap #pip install geemap
import pandas as pd
import time

# Trigger the authentication flow.
ee.Authenticate()

# Initialize the library.
ee.Initialize(project='cnnafrica-420018')

coords = pd.read_csv('C:/Users/Luis/OneDrive/Desktop/CNN_Africa/data/data/bounding_boxes_vertical.csv')
for i in range(len(coords)):

    # if i <=25:
    #     continue

    # if i == 10 or i == 11:
    #     print(f"Skipping iteration {i+1}...")
    #     continue

    box = coords.loc[i, :].values.flatten().tolist()

    geometry = ee.Geometry.Rectangle(box)
    name = "2019_africa_landcover_" + str(i+1)

    dataset = ee.Image('COPERNICUS/Landcover/100m/Proba-V-C3/Global/2019').select(
        'discrete_classification'
    ).clip(geometry)

    projection = dataset.projection().getInfo()

    task = ee.batch.Export.image.toDrive(image=dataset,
                                        description=name,
                                        fileNamePrefix=name,
                                        crs= projection["crs"],
                                        crsTransform = projection["transform"],
                                        region = geometry,
                                        maxPixels= 1e10,
                                        fileFormat='GeoTIFF')
    task.start()

    while task.status()['state'] != 'COMPLETED':
        print(f"Task {name} is {task.status()['state']}. Waiting for completion...")
        time.sleep(10)
    
    print(f"Task {name} completed successfully!")
```
